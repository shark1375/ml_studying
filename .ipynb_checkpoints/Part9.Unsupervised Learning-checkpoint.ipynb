# 9.1. Clustering
- 이번 챕터에서는 KMeans와 DBSCAN을 활용할 것
- 군집화는 비슷한 샘플을 구별하여 하나의 클러스터에 할당하는 작업임
- 분류와 마찬가지로 각 샘플은 하나의 그룹에 할당되지만, 비지도 학습임
- 주로 사용되는 분야는 아래와 같음
    - 1) 고객 분류 -> 추천시스템을 만들기위한 기반
    - 2) 데이터 분석 -> 데이터를 특성별로 클러스터링해내어 따로따로 분석하면 도움이 됨
    - 3) 차원축소 -> 각 클러스터에 대한 친화성을 측정할 수 있음
    - 4) 이상치 탐지 -> 모든 클러스터에 대한 친화성이 낮은 샘플은 이상치일 가능성이 높음
        - FDS에서 흔히 활용
    - 5) 준지도학습 -> 레이블된 샘플이 적을 때 사용 -> 레이블별로 특성을 파악하고, 해당 레이블을 역전파
    - 6) 검색엔진 -> 구글 이미지 같은 것
    - 7) 이미지 분할 -> 색 기반 픽셀 단위로 클러스터링 -> 각 픽셀을 해당 클러스터의 평균색으로 변경
        - 물체 윤곽을 확실히 구분할 수 있게 되어, 탐지/추적에 흔히 활용
- 보편적인 정의는 없으며, 상황에따라 다른 알고리즘, 다른 결과물을 도출함
    - 어떤 알고리즘은 centroid라 부르는 포인트를 중심으로 샘플을 찾음
    - 어떤 알고리즘은 샘플이 밀집되어있는 연속된 영역을 찾음
    - 어떤 알고리즘은 어떠한 모야을 가가져도 상관이 없음
    - 어떤 알고리즘은 hierarchical하게 클러스터링을 함
- 가장 유명한 KMeans와 DBSCAN을 살펴보고 비선형 차원축소, 준지도학습, 이상치 탐지와 같은 애플리케이션 확인

## 9.1.1. K-Means Clustering

- 우선 데이터셋 생성

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs

blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]])
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

X, y = make_blobs(n_samples=2000, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)


def plot_clusters(X, y=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1)
    plt.xlabel("$x_1$", fontsize=14)
    plt.ylabel("$x_2$", fontsize=14, rotation=0)
    
plt.figure(figsize=(8, 4))
plot_clusters(X)
plt.show()
```

- 사이킷런의 cluster패키지 내 KMeans클래스 활용
- KMeans(n_cluster)
    - 알고리즘이 찾을 클러스터갯 k를 지정해야함
        - 위 그래프에서는 다섯개라는걸 쉽게 파악할 수 있음
    - 일반적으로는 이걸 찾는 게 쉬운 일이 아님
    - fit_predict메서드로 로 한 번에 결과물 까지 도출할 수 있음
        - labels_인스턴스에 훈련된 샘플의 레이블을 갖고 있음
        - cluster_centers_인스턴스에 센트로이드 지점도 확인할 수 있음
    - 클러스터의 결정경계를 그려보면 보로노이 다이어그램을 얻을 수 있음
        - 평면을 특정 점 까지의 거리가 가장 가까운 점의 집합으로 분할한 그림

```python
from sklearn.cluster import KMeans

k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
y_pred = kmeans.fit_predict(X)

y_pred == kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
X_new = np.array([[0, 2], [3, 2], [-3, 3], [-3, 2.5]])
kmeans.predict(X_new)
```

- KMeans클러스터링 알고리즘의 경우, 클러스터의 크기가 많이 다르면 잘 작동하지 않음
    - 마치 우리 회사의 Imbalanced Dataset과 같음
    - 알고리즘이 매우 단순한 탓

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=30, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=50, linewidths=50,
                color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
plt.figure(figsize=(8, 4))
plot_decision_boundaries(kmeans, X)
plt.show()
```

- 하드 군집(Hard Clustering) : 하나의 샘플을 하나의 클러스터에 할당하는 것
- 소프트 군집(Soft Clustering) : 각 클러스터마다 샘플에 점수를 부여하는 방식
    - 이 점수는 유사도 점수(친화성 젓무)가 될 수 있음
    - KMeans클래스 내 transform()메서드를 활용하면 각 샘플과 각 센트로이드 사이의 거리를 반환함
    - 이 방식은 차원축소의 방법 중 한 가지로 활용될 수도 있음

```python
kmeans.transform(X_new)
```

### 9.1.1.1. K-Means Algorithm
- 
    - 처음에 기준이될 센트로이드를 무작위로 선정함(k개)
    - 그 다음 샘플에 레이블을 할당하고, 센트로이드 업데이트하는 걸 반복
        - 센트로이드가 변화가 없을 때 까지 계속함
    - 일반적으로 이 알고리즘은 제한된 횟수 안에 수렴하는 것을 보장함(매우 작음)
        - 최적화하는 것이 기본으로 잡혀있기 때문에 따로 hyperparametr에 대한 튜닝이 필요 없음
    - 계산 복잡도는 샘플개수 m, 클러스터 k, 차원 개수 n에 선형적이며, 가장 빠른 군집 알고리즘임
    - 다만, 어느 정도 선에서 알고리즘이 수렴하는 것은 확실하지만, 최적의 군집으로 모이지는 않을 수 있음
        - 따라서 센트로이드 초기화를 개선하여 이런 위험을 최대한 줄일 필요가 있음

### 9.1.1.2. 센트로이드 초기화
- 
    - 만일 센트로이드 위치를 근사하게 알 수 있는 방법이 있다면, KMeans함수 인자 중 init 매개변수에 초기 센트로이드 리스트를 담은 넘파이 배열을 지정하고 n_init을 1로 설정하여 추가 업데이트를 막을 수 있음

```python
good_init = np.array([[-3, 3], [-3, 2], [-3, 1], [-1, 2], [0,2]])
kmeans = KMeans(n_clusters = 5, init = good_init, n_init = 1)
y_pred = kmeans.fit_predict(X)
```

- 
    - 또 다른 방법은 랜덤 초기화를 다르게 하여 여러 번 알고리즘을 실행하고 가장 좋은 솔루션을 선택하는 것
        - 랜덤 초기화 횟수는 n_init 매개변수로 조절함
            - 디폴트는 10으로 잡힘
            - 즉, fit() 메서드 호출 시 전체 알고리즘이 10회 실행된다는 뜻
        - 사이킷런은 지정된 횟수만큼 반복된 결과물 중 최선의 솔루션을 반환함
    - 이 때, 최선의 솔루션을 판단하는 성능지표가 이너셔(Inertia)임
        - 이너셔는 각 샘플과 가장 가까운 센트로이드 사이의 평균 제곱거리를 의미함
    - KMeans클래스는 알고리즘을 n_init번 실행하고, 이너셔가 가장 낮은 모델을 반환함
    - KMeans인스턴스의 inertia_ 변수를 통해 이너셔 확인 가능
        - fit이나 fit_prediction을 통해서 일단 인스턴스를 생성해야함

```python
kmeans.inertia_
```

- 
    - KMeans 클래스 인스턴스의 inertia_는 최적의 거리를 보여주는 것(작을수록 좋은)
    - 그러나 score() 메서드는 이너셔의 음수값을 반환
        - sklearn의 경우 score는 큰 값이 좋은 것이라다는 규칙이 있기 때문
        - 이너셔는 작을 수록 좋으니까, 음수로 변환하여, 해당 규칙을 따르게 한 것

```python
kmeans.score(X)
```

- 
    - 2006년에 데이비드 아서와 세르게이 바실비츠키가 K-means++ 알고리즘을 제안함
        - 이 알고리즘은 다른 센트로이드와 거리가 먼 센트로이드를 선택하는 개선된 초기화 단계를 제안
        - 이 방법은 k means알고리즘이 최적이 아닌 솔루션으로 수렴할 가능성을 크게 낮춤
    - 최적의 솔루션을 찾기위해 실행하게 되는 추가적인 알고리즘 반복횟수를 크게 줄일 수 있었음
    - 원리는 다음과 같음
        - 1) 데이터에서 무작위로 균등하게 하나의 센트로이드 c(1)을 지정
        - 2) D(x(i))^2/Sigma(j=1, m)[D(x(j))^2]의 확률로 샘플 x(i)를 새로운 센트로이드 c(i)로 선택
            - 이 때 D(x(i))는 샘플 x(i)와 이미 선택된 가장 가까운 센트로이드까지의 거리
            - 이 확률 분포는 이미 선택한 센트로이드에서 멀리 떨어진 샘플을 다음 센트로이드로 선택할 가능성을 높이는 역할을 함
        - 3) k개의 센트로이드가 선택될 때까지 반복
    - sklearn의 KMeans클래스는 k-means ++ 알고리즘의 초기화방법을 디폴트로 사용함
        - 원래 방식을 사용하고 싶다면 init변수를 "random"으로 지정하면 됨

### 9.1.1.3. kmeans 속도 개선 및 min-batch kmeans
- 
    - 13년 찰스 엘칸의 논문에서 새로운 개선안 제안
    - 불필요한 거리 계산을 피하며 알고리즘의 속도를 올릴 수 있는 방안
        - 삼각 부등식(triangle inequality)를 활용
            - 두 점 사이의 직선은 항상 가장 짧은 거리가 되도록 함(AC <= AB + BC)
        - 또한, 샘플과 센트로이드 사이의 거리를 위한 하한선과 상한선을 결정하여 유지함
    - KMeans클래스에서 기본으로 활용하는 알고리즘임
        - 원래 알고리즘을 활용하려면, KMeans내 algorithm인자를 full로 지정하면 되긴하는데...굳이...?
- 
    - 10년도에는 스컬리의 논문에서 새로운 k-means 알고리즘이 제시됨
    - 전체 데이터셋을 한 번에 이용하는 것이 아니라, 각 반복마다 미니배치를 활용하여 센트로이드를 조금씩 이동하도록 함
        - 일반적으로, 3배~4배정도 속도가 올라감
        - 또한, 메모리에 들어가지 않는 대량의 데이터 셋에도 클러스터링 알고리즘을 적용할 수 있음
    - sklearn의 MiniBatchKMeans클래스에 이 알고리즘이 구현되어있으며, 사용방식은 크게 다르지 않음

```python
from sklearn.cluster import MiniBatchKMeans

minibatch_kmm = MiniBatchKMeans(n_clusters = 5)
minibatch_kmm.fit(X)
```

- 
    - 데이터셋이 메모리에 들어가지 않을 때 활용할 수 있는 좋은 방법
    - 하지만 더 간단한 방법은 memmap클래스를 활용하는 것
    - 또는, MiniBatchKMeans 클래스의 partial_fit()메서드에 한 번에 하나의 미니배치를 전달하는 것
        - But 초기화 횟수가 증가하고, 직접 최적의 결과를 골라야하는 경우가 많아, 다소 어려움
- 
    - 미니배치가 일반적으로, 일반 KMeans보다는 빠름
        - 그러나 이너셔는 일반적으로 더 나쁨
        - 특히, 클러스터의 개수가 증가할 때 그러함

### 9.1.1.4. Finding Optimal #Cluster
- 
    - 일반적으로 k값을 찾는것은 매우 어려운 문제임
    - 잘못된 클러스터 개수를 선택해버릴 경우
        - k가 작으면 별개의 클러스터를 합치고, 크면 하나의 클러스터를 여러개로 쪼개는 것을 알 수 있음
    - 가장 이너셔가 작은 모델을 선택할 수는 있으나, 그렇게 간단한 차원의 문제가 아님
        - k가 커질 수록 이너셔는 당연히 작아질 수밖에 없음
        - 따라서 이너셔가 작다/크다는 클러스터링이 잘 되었는지 안되었는지에 대한 판단 지표는 될 수 있지만, k를 결정하는 절대 지표가 될 수 없음
    - 이 때 탐색을 할 필요가 있는 것이 엘보우 포인트임

```python
kmeans_k3 = KMeans(n_clusters=3, random_state=42)
kmeans_k8 = KMeans(n_clusters=8, random_state=42)
kmeans_k3.fit(X)
kmeans_k8.fit(X)

plot_decision_boundaries(kmeans_k3, X)
plt.show()
plot_decision_boundaries(kmeans_k8, X)
plt.show()
```

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(X)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]

plt.figure(figsize=(8, 3.5))
plt.plot(range(1, 10), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.annotate('Elbow',
             xy=(4, inertias[3]),
             xytext=(0.55, 0.55),
             textcoords='figure fraction',
             fontsize=16,
             arrowprops=dict(facecolor='black', shrink=0.1)
            )
plt.axis([1, 8.5, 0, 1300])
plt.show()
```

```python
plot_decision_boundaries(kmeans_per_k[4-1], X)
plt.show()
```

- 
    - k에 대한 정답을 몰랐다면 k = 4가 꽤 좋은 선택이 되었을 것
    - 이보다 작은 값은 변화가 심하고, 더 큰 값은 도움이 되지 않음
- 
    - 그런데 이렇게 엘보우 포인트를 찾아서 클러스터 개수를 선정하는 방식은 너무 엉성함
    - 더 정확한 방법은 또 따로 있음
- 실루엣 점수(silhouette score)
    - 모든 샘플에대한 실루엣 계수의 평균
    - 각 샘플에 대한 실루엣 계수는 (b-a)/(max(a,b))로 계산함
        - a : 동일 클러스터 내 다른 샘플까지의 평균 거리
        - b : 가장 가까운 클러스터까지의 평균 거리
    - 실루엣 계수는 -1 ~ 1 까지 바뀔 수 있으며, 1에 가까울 수록 자신의 클러스터 내에 잘 속해있고, 타 클러스터와는 멀리 떨어져있다는 의미
    - 0에 가까우면 경계에 위치
    - -1에 가까우면 잘못된 클러스터에 할당되었다는 의미
    - 이는 사이킷런의 silhouette_score()함수로 계산할 수 있음

```python
from sklearn.metrics import silhouette_score
silhouette_score(X, kmeans.labels_)
```

```python
silhouette_scores = [silhouette_score(X, model.labels_)
                     for model in kmeans_per_k[1:]]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.axis([1.8, 8.5, 0.55, 0.7])
plt.show()
```

- 
    - 실루엣 계수를 보니 k = 5도 꽤나 괜찮은 선택이었다는 것을 알 수 있음
    - k = 6, 7에 비하면 4가 굉장히 좋은 선택이었음
        - 위 두 가지 사실은 이너셔 그래프로는 보기 어려웠음
- 실루엣 다이어그램(silhouette diagram)
    - 모든 샘플의 실루엣 계수를 할당된 클러스터와 계숫값으로 정렬하여 그리면 더 많은 정보를 담은 그래프를 얻을 수 있음
        - 그래프의 높이 : 클러스터가 포함하고 있는 샘플의 개수
        - 그래프이 너비 : 이 클래스에 포함된 샘플의 정렬된 실루엣 계수
            - 결과적으로 넓이가 넓을 수록 좋음
    - 그리고 수직 ㅍ파선은 각 클러스터 개수에 해당하는 실루엣 점수임
        - 한 클러스터의 샘플 대부분이 이 점수보다 낮은 계수를 가지면(즉, 대부분의 샘플이 파선 왼쪽에 있으면), 클러스터의 샘플이 다른 클러스터와 너무 가깝다는 것을 의미함
        - 즉, 해당 클러스터는 나쁜 클러스터라는 것
    - 또한, k = 4일때는 하나의 클러스터가 매우 큼
        - 클러스터 크기는 균일하게 가는 게 좋으므로, 실루엣 스코어에서는 4가 다소 앞서나, 5를 선택하는 것이 좋음

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter
import matplotlib as mpl

plt.figure(figsize=(11, 9))

for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)
    
    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(X, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")
    
    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)

plt.show()
```

- 예를들어,
    - 50개 훈련 세트를 클러스터로 모으고, 각 클러스터에서 센트로이드에 가장 가까운 이미지를 찾아보자
        - 대표 이미지(representative Image)

```python
k = 50
kmeans = KMeans(n_clusters = k)
X_digits_dist = kmeans.fit_transform(X_tr)
representative_digit_idx = np.argmin(X_digits_dist, axis = 0)
X_representative_digits = X_tr[representative_digit_idx]
```

```python
plt.figure(figsize = (20, 20))
plt.imshow(X_representative_digits, cmap = "gray", interpolation = "nearest")
plt.show()
```

- 위 이미지를 보고, 수동으로 값을 할당해볼 수 있음
- 하지만 잘 안보이니...
    - 책에서는 이미지 보고 직접 할당시킴

```python
y_representative_digits = y_tr[representative_digit_idx]
```

```python
y_representative_digits
```

- 어쨌든 위의 값들은 무작위가 아닌, 해당 클러스터들을 대표하는 이미지임

```python
log_reg = LogisticRegression()
log_reg.fit(X_representative_digits, y_representative_digits)
log_reg.score(X_t, y_t)
```

- 매우 성능이 좋아짐
- 고작 50개로만 진행했을 뿐인데도
- 하지만 한 단계 더 나아갈 수 있음
    - 레이블 전파
    - 같은 클러스터 내에 대표값으로 레이블을 부여하는 것

```python
y_train_propagated = np.empty(len(X_tr), dtype = np.int32)
for i in range(k):
    y_train_propagated[kmeans.labels_==i] = y_representative_digits[i]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_tr, y_train_propagated)
log_reg.score(X_t, y_t)
```

- but 지금은 문제가 있는게, 동일 클러스터 전체에 전파하는 문제
    - 즉, 센트로이드와 멀리 떨어져 각 클러스터와 클러스터 사이의 경계에 있는 놈들도 똑같이 값이 부여됨
- 이번엔 센트로이드와 가까운 샘플의 20%만 전파하고 학습시켜보자

```python
percentile_closest = 20

X_cluster_dist = X_digits_dist[np.arange(len(X_tr)), kmeans.labels_]
for i in range(k):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1
    
partially_propagated = (X_cluster_dist != -1)
X_tr_partially_propagated = X_tr[partially_propagated]
y_tr_partially_propagated = y_tr[partially_propagated]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_tr_partially_propagated, y_tr_partially_propagated)
log_reg.score(X_t, y_t)
```

##### ACTIVE LEARNING
- 모델과 훈련세트를 지속적으로 향상시키기 위해 능동학습을 다수 반복할 수 있음
- 여러 전략이 많으나, 가장 널리 사용되는 능동학습 방법론은 불확실성 샘플링(Uncertainty Samplinmg)임
- 작동 방식은 다음과 같음
    - 1) 지금까지 수집한 레이블된 샘플 -> 모델링
        - 이 모델로 unlabeled 샘플에 대한 label예측을 만들어냄
    - 2) 모델이 가장 불확실하게 예측한 샘플을 전문가들에게 보내 레이블을 붙임
    - 3) 레이블을 부여하는 노력만큼의 성능이 향상되지 않을 때까지 반복
- 다른 방식
    - 모델을 가장 크게 바꾸는 영향력 높은 샘플이나, 모델의 검증점수를 가장 크게 떨어뜨리는 샘플, 여러 개의 모델이 동일한 예측을 내지 않는 샘플에 대하여 레이블 요청

## 9.1.6. DBSCAN
- 밀집된 연속지역을 클러스터로 정의
- 작동 방식은 다음과 같음
     - 1) 알고리즘이 각 샘플에서 작은 거리인 epsilon내에 샘플이 몇개 있는지 셈
         - 이 지역을 epsilon neighborhood(e-neighbor)라고 부름
     - 2) 자신을 보함하여 e이웃 이내 적어도 min_sample개의 샘플이 있다면 이를 core instance(핵심샘플)로 간주
     - 3) core instance의 이웃에 있는 모든 샘플은 동일 클러스터로 간주
         - 이웃에는 다른 core instance가 들어올 수 있음
         - 따라서, 핵심 샘플의 이웃 - 의 이웃은 계속해서 하나의 클러스터를 형성
     - 4) 핵심 샘플도 아니고 이웃도 아닌 샘플은 outlier로 판단
         - 오...이러면 FDS등에 쓰일 수 있을 듯
 - 이 알고리즘은
     - 1) 모든 클러스터가 충분히 밀집되어있고
     - 2) 밀집 지역과 밀집되지 않은 지역이 잘 구분될 떄 좋은 성능을 냄
 - 사이킷런의 DBSCAN클래스는 예상대로 사용법이 간단함

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 1000, noise = 0.05)
dbscan = DBSCAN(eps = 0.05, min_samples = 5)
dbscan.fit(X)
```

```python
set(dbscan.labels_)
```

- -1은 DBSCAN모델이 해당 샘플을 이상치로 판단했다는 의미임
- 핵심 샘플의 인덱스는 인스턴스 변수 core_sample_indices_에서 확인할 수 있음
- 핵심 샘플 자체는 components_에 저장되어있음

```python
len(dbscan.core_sample_indices_)
```

```python
dbscan.components_
```

- epsilon값을 0.2로 증가시킨 모델과 시각화하여 비교

```python
dbscan2 = DBSCAN(eps=0.2)
dbscan2.fit(X)
```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    plt.scatter(non_cores[:, 0], non_cores[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
```

```python
plt.figure(figsize=(9, 3.2))

plt.subplot(121)
plot_dbscan(dbscan, X, size=100)

plt.subplot(122)
plot_dbscan(dbscan2, X, size=600, show_ylabels=False)

plt.show()
```

- epsilon값이 0.05로 많이 낮았을 떄는 군집을 너무 많이 쪼개고, 많은 샘플을 이상치로 판단함
- epsilon값을 0.2로 높이자, 완전한 군집을 얻게 됨(dbscan2)

- dbscan모델의 경우 predict를 제공하지 않고 fit_predict()메서드를 제공
    - 즉 새로운 샘플에 대해 기존 알고리즘을 토대로한 클러스터 예측이 불가능
    - 새로운 샘플이 들어올 때마다 새로운 모델 피팅이 필요함
- 그 이유는, 이렇게 clustering을 해놓은 경우, 다른 classification알고리즘을 도입해서 분류작업을 하면 해당 작업을 더 잘 수행할 수 있기 때문임
    - 그러므로 사용자가 필요한 예측기를 선택해야함
        - ex) KNeighborsClassification

## 9.1.7. 기타 군집 알고리즘

### 9.1.7.1. 병합 군집(agglomerative clustering)
- 클러스터 계층을 밑바닥부터 위로 쌓아 구성
- 처음엔 하나의 샘플에서 시작하여 서로 인접한 클러스터 쌍을 엮어나감
- 결국 병합된 클러스터 쌍을 트리로 모두 그리면 클러스터의 이진 트리를 얻을 수 있음
- 시작이 각기 샘플이기 떄문에 결국 각 리프는 개별 샘플임
- 다만, 연결 행렬이 없으면 대규모 데이터셋으로 확장할 수 없음
    - 연결 행렬은 이웃 샘플간의 거리를 담은 mxm의 sparse matrix
    - 예를 들어 sklearn.neighbors.kneighbors_graph() 반환값
    
### 9.1.7.2. BIRCH
- Balanced Iterative Reducing and Clustering using Hierarchies
- 대규모 데이터셋에 적용하기 위해 고안된 알고리즘
- 특성 개수가 너무 많지 않다면(20개 이하), 배치 k-means보다 빠르고 비슷한 결과를 만들어냄
- 훈련 과정에서 새로운 샘플을 각기 클러스터에 빠르게 할당할 수 있는 트리 구조를 생성해냄
- 다만, 트리에 모든 샘플을 저장하고 있지 않기 때문에, 메모리 문제를 낮추고, 대용량 데이터셋을 다룰 수 있게됨

### 9.1.7.3. 평균이동(Mean-shift)
- 각 샘플을 중심으로 하는 원을 그리고, 각 원마다 원 안에 포함된 모든 샘플의 평균을 구함
- 그리고 원의 중심을 평균점으로 이동
- 모든 원이 움직이지 않을 때까지 평균이동을 반복함
- 지역의 최대 밀도를 찾을 때까지 높은 밀도의 방향으로 원을 이동시킴
- 동일 지역에(or 충분히 가깝게) 안착한 원 안에 있는 모든 샘플은 하나의 클러스터로 엮이게됨
- DBSCAN과 비슷한 특징
    - 1) 모양이나 개수에 상관없이 클러스터 서칭 가능
    - 2) 하이퍼파라메터의 수가 적음(bandwidth라고 불리는 "원 반경" 딱 하나)
    - 3) 국부적인 밀집도 추정에 의존
        - 하지만, 클러스터 내부 밀집도가 불균형할 때 여러개로 클러스터를 쪼개는 경향
- O(m^2)의 계산복잡도를 가지고 있어서 대규모 데이터셋엔 적합하지 않음

### 9.1.7.4. 유사도 전파(Affinity Propagagtion)
- 투표 방식을 이용함
- 샘플은 자신을 대표할 수 있는 비슷한 샘플에 투표를 실시함
- 알고리즘이 수렴하면 각 대표와 투표한 샘플이 클러스터를 형성
- 크기가 다른 여러개의 클러스터를 감지할 수 있다는 장점
    - 우리 회사 데이터 세에는 이게 더 잘맞을 수도 있을 듯
- O(m^2)의 계산복잡도 -> 따라서 대규모 셋에는 잘 안맞음

### 9.1.7.5. 스펙트럼 군집(Spectral Clustering)
- 샘플 사이의 유사도 행렬을 받아 저차원 임베딩을 만듦(i.e. 차원축소 실행)
- 그 다음 이 저차원 공간에서 또 다른 군집 알고리즘 활용(사이킷런에서는 저차원 임베딩 후 kmeans실행)
- 복잡한 클러스터 구조를 감지하고 그래프 컷을 찾는데 사용할 수 있음
- 샘플개수가 많거나, 클러스터의 크기 차이가 매우 크면, 잘 작동하지 않음

# 9.2. Gaussian Mixture Model(GMM)

- 샘플이 파라미터가 알려지지 않은 여러 개의 혼합된 가우시안 분포에서 생성되었다고 가정하는 확률 모델
- 하나의 가우시안 분포에서 생성되었다고 가정되는 샘플들 -> 하나의 클러스터를 형성
    - 일반적으로 이 클러스터는 타원형
- 각 클러스터는 타원의 모양, 크기, 밀집도, 방향이 다름
- 우리가 아는 것은 샘플이 주어지면 이 샘플이 어떤 건지는 모르지만 어떠한 가우시안 분포로부터 생성되었다는 사실뿐
    - 이 가우시안 분포가 어떤 분포인지, 이 분포의 파라미터는 무엇인지 모르는 상태임

- GMM에는 여러 변종이 있는데, 가장 간단한 것은 GussianMixture클래스에 구현되어있음
    - 여기에서는 사전에 가우시안 분포의 개수 k를 알아야함
        - 약간 kmeans와 비슷하네
        - 가우시안 분포의 개수는 곧 클러스터의개수임
    - 데이터셋 X가 다음과 같은 확률 과정을 통해 생성되었다고 가정해보자
        - 1) 샘플마다 k개의 클러스터에서 랜덤하게 한 클러스터가 선택됨
        - 2) j번째 클러스터를 선택할 확률은 클러스터의 가중치 pi(j)로 정의함
        - 3) i번째 샘플의 클러스터 인덱스는 z(i)로 표시함
        - 4) z(i) = j, 즉 i번째 샘플이 j클러스터에 할당된다면, 이 샘플의 위치 x(i)는 평균이 mu(j)이고 공분산행렬이 Sigma(j)인 가우시안 분포에서 랜덤하게 샘플링 됨
        - 5) 이를 수식으로 표기하면, x(i) ~ N(mu(j), Sigma(j))로 씀
    - 이러한 생성 과정은 그래프모형으로 나타낼 수 있음
        - 확률 변수사이의 조건부 의존성이 드러남
        - 
        - 위 그래프를 보는 방법은 다음과 같음
            - 1) 원은 확률변수를 나타냄
            - 2) 사각형은 고정값, i.e. 파라미터를 의미함
            - 3) 큰 사각형을 plate라고 부름 -> 이 사각형 안의 내용이 여러 번 반복됨을 의미함
            - 4) 각 플레이트 오른쪽 아래의 숫자는 플레이트의 내용이 얼마나 반복되는지 표시
                - 따라서 m개의 확률변수 z(i)와 m개의 x(i)가 있음
                - 또한 k개의 평균 mu(j)와 k개의 공분산 Sigma(j)rk dlTdma
                - 마지막으로 하나의 가중치 벡터 pi가 있음
            - 5) 각 변수 z(i)는 가중치 pi를 갖는 범주형 분포에서 샘플링 됨
                - 각 변수 x(i)에 해당하는 클러스터 z(i)로 정의된 평균과 공분산 행렬을 활용하여 정규분포로부터 샘플링됨
            - 6) 실선 화살표는 조건부 의존성을 의미함
                - 예를 들어 각 확률변수 z(i)의 확률분포는 가중치 벡터 pi에 의존함
                - 화살표가 플레이트 경계를 가로지르면 해당 플레이트의 모든 반복에 적용한다는 의미
                - 예를 들어 pi는 x(1) ~ x(m)까지의 모든 확률 변수의 분포에 필요조건임
            - 7) z(i) ~ x(i)까지의 구불구불한 화살표 = 스위치
                - z(i)값에 따라 x(i)가 다른 가우시안 분포에서 샘플링 될 것
                - ex) z(i) = j -> x(i) ~  N(mu(j), Sigma(j))
            - 8) 색이 채워진 원 = 알려진 값
                - 따라서 이 경우에는 x(i)만 알고 있는 값
                - 이를 observed variable이라고 함
                - 알려지지 않은 값 z(u)는 latent variable이라고 부름

- 우선 이 모델로, X가 주어졌을 때 가중치 pi와 전체 분포의 파라미터 mu(1) ~ mu(k), Sigma(1) ~ Sigma(k)까지를 추정할 수 있음

- 이 추정 과정은 기댓값-최대화(Expectation Maximization_EM) 알고리즘에 기반함
- 이 알고리즘은 kmeans와 공통점이 많음
    - 1) 클러스터 파라미터를 랜덤하게 초기화하고 수렴할 때까지 두 단계를 반복함
    - 2) 1단계 : 기댓값 단계
        - 샘플을 클러스터에 할당
    - 3) 2단계 : 최대화 단계
        - 클러스터 업데이트
    - 상당히 비슷함.
    - 군집 입장에서 바라보면, EM을 클러스터 중심(mu(1) ~ mu(k))뿐만 아니라 크기, 모양, 방향(Sigma(1) ~ Sigma(k))과 클러스터의 상대적 가중치(pi(1) ~ pi(k))를 찾는 kmeans의 일반화된 모델이라고 생각할 수 있음
- 다만 kmeans와 달리 소프트 클러스터 할당을 활용함(kmeans는 하드 클러스터 할당)
    - 기댓값 단계에서 현재의 클러스터 파라미터에 기반하여 각 클러스터에 속할 확률을 예측
    - 최대화 단계에서 각 클러스터가 데이터셋에 있는 모든 샘플을 활용하여 업데이트
    - 클러스터에 속할 추정확률로 샘플에 가중치가 적용됨
        - 이 추정확률을 Responsibility라고 부름
    - 최대화 단계에서 클러스터 업데이트는 책임이 가장 많은 샘플에 큰 영향을 받음
- 주의할 점
    - kmeans와 비슷한만큼, 역시 안좋은 해답을 내놓을 가능성이 있음

- 클러스터별 평균, 결정경계(파선), 밀도 등고선을 그려보자

```python
from matplotlib.colors import LogNorm

def plot_gaussian_mixture(clusterer, X, resolution=1000, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = -clusterer.score_samples(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z,
                 norm=LogNorm(vmin=1.0, vmax=30.0),
                 levels=np.logspace(0, 2, 12))
    plt.contour(xx, yy, Z,
                norm=LogNorm(vmin=1.0, vmax=30.0),
                levels=np.logspace(0, 2, 12),
                linewidths=1, colors='k')

    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z,
                linewidths=2, colors='r', linestyles='dashed')
    
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)
    plot_centroids(clusterer.means_, clusterer.weights_)

    plt.xlabel("$x_1$", fontsize=14)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
plt.figure(figsize=(8, 4))
plot_gaussian_mixture(gm, X)
plt.show()
```

                - 1) spherical : 모든 클러스터가 원형(분산은 다를 수 있어, 지름이 달라질 수 있음)
                - 2) diag : 어떤 타원형도 가능하나 타원의 축은 좌표축과 나란해야함(즉, 공분산 행렬이 대각행렬)
                - 3) tied : 모든 클러스터가 동일한 타원 모양, 크기, 방향을 가짐(즉, 공분산 행렬이 동일함)
            - 디폴트는 full : 모양/크기/방향의 제약 X

```python
gm_full = GaussianMixture(n_components=3, n_init=10, covariance_type="full", random_state=42)
gm_tied = GaussianMixture(n_components=3, n_init=10, covariance_type="tied", random_state=42)
gm_spherical = GaussianMixture(n_components=3, n_init=10, covariance_type="spherical", random_state=42)
gm_diag = GaussianMixture(n_components=3, n_init=10, covariance_type="diag", random_state=42)
gm_full.fit(X)
gm_tied.fit(X)
gm_spherical.fit(X)
gm_diag.fit(X)
```

```python
def compare_gaussian_mixtures(gm1, gm2, X):
    plt.figure(figsize=(9, 4))

    plt.subplot(121)
    plot_gaussian_mixture(gm1, X)
    plt.title('covariance_type="{}"'.format(gm1.covariance_type), fontsize=14)

    plt.subplot(122)
    plot_gaussian_mixture(gm2, X, show_ylabels=False)
    plt.title('covariance_type="{}"'.format(gm2.covariance_type), fontsize=14)
```

```python
compare_gaussian_mixtures(gm_tied, gm_spherical, X)
plt.show()
```

```python
compare_gaussian_mixtures(gm_full, gm_diag, X)
plt.tight_layout()
plt.show()
```

## 9.2.2. 클러스터 개수 선택
- K-means에서는 이너셔나 실루엣 점수를 사용하여 클러스터 개수를 선택했음
- 가우시안 혼합에서는 이런 지표를 사용할 수가 없음
    - 위 지표들은 클러스터가 타원형이거나,크기가 다를 때는 안정적이지 않음
- 대신 BIC(Bayesian Information Criterion)나 AIC(Akaike Information Criterion)와 같은 이론적 정보기준(Theoretical Information Criterion)을 최소화하는 모델을 찾음
    - BIC = log(m)p - 2log(L_hat)
    - AIC = 2p - 2log(L_hat)
        - m : # Samples
        - p : # Parameter
        - L_hat : 모델의 가능도 함수(Likelihood_function)의 최댓값
- BIC와 AIC 모두 학습할 파라메터가 많은(클러스터가 많은) 모델에 벌칙을 가하고, 데이터에 더 적합한 모델에 보상을 더함
- 종종 AIC와 BIC는 동일한 모델을 선택함
    - 다를 경우, BIC가 선택한 모델이 AIC모델보다 간단한 경향이 있음
    - But 데이터가 아주 잘 맞지 않을 수 있음(특히, 데이터가 커질 때)

##### 가능도함수
- 확률과 가능도는 "통계학"에서는 다른 의미로 활용됨(비슷한 기능 but 아예 다른 개념)
    - 1. 확률 : 파라메터 theta인 확률모델이 주어지면(즉, theta에 대해 알고 있다면), 확률은 미래 출력 x가 얼마나 그럴듯한지에 대한 설명
    - 2. 가능도 : 가능도는 반대로 출력 x를 알고 있을 때 theta가 얼마나 그럴듯한지에 대해 설명
- 예를 들어, -4와 +1이 중심인 두 개의 가우시안 분포를 가진 ID 혼합 모델 생각
    - 이 모델은 두 분포의 표준편차를 제어하기 위한 theta를 하나 가짐
    - f(x:theta)가 있다고 가정
        - 미래 출력 x의 확률분포를 예측하려면 theta를 지정할 필요가 있음
            - 예를 들어 theta = 1.3이라면, f(x:theta = 1.3)에 해당하는 확률분포를 얻게 됨
            - 여기서 x가 -2와 2 사이에 들어갈 확률을 예측하고 싶다면 이 범위에서 pdf에 대한 적분을 해야함
        - 하지만, theat를 모르는 상황에서 샘플 x = 2.5를 추정했다면?
            - 이 때는 가능도 함수 L(theat|x=2.5) = f(x = 2.5:theta)를 얻게 되는 것
    - 다시 말해서, pdf는 x의 함수로, theta값을 고정하고 생각함
    - 반면 Likelihood는 theata의 함수임
        - 즉, pdf가 아님!
        - 그래서 pdf는 전체 범위에 대해 적분 시 1이 되지만, L을 모든 theta에 대해 적분하면 어떤 값이든 나올 수 있음(양수 범위에서)
- 근데, 보통은 pdf도 모르고 theta에 대해서도 모르는 케이스가 다반사
    - 그러므로, X가 주어진 상황에서 theta를 추측해내기 위해 가능도함수를 최대로 하는 값을 찾아내야함
    - 그게 MLE(Maximum Likelihood Estimation) : 최대가능도 추정
        - theta_hat = 1.5가정
        - 만약 theta에 대한 Prior Probability가 존재한다면, L(theta|x)g(theta)를 최대화하는 것도 가능
            - Maximum a-posteriori(MAP) = L(theta|x)g(theta)
        - MAP은 파라메터 값을 제약하므로, MLE의 규제버전이라고 생각하면 됨
- 일반적으로는 Likelihood자체를 키우는 것보다는 log_likelihood를 최대화하는 것이 더 쉬움
    - 그래서 보통 Likelihood보다는 log_likelihood를 최대화하여 theata를 구하게 됨
- theta_hat을 찾았다면, AIC/BIC를 찾기 위한 값인 L_hat = L(theta_hat, X)를 계산할 수 있게됨

- 가우시안모델 클래스에 bic와 aic메서드를 이용하여 데이터 X에 적합하면, 해당 값들을 구할 수 있음
    - 두 값은 모두 가장 작은 것을 찾아야함

```python
print(gm.bic(X))
print(gm.aic(X))
```

    - 지금 Bayesian GMM에서는 0~1범위임
    - 예를 들어 SBP(Stick Breaking Process)
        - PI = [0.3, 0.6, 0.5, ...]을 가정
            - 샘플의 0.3이 클러스터 0에 할당됨
            - 남은 샘플의 0.6이 클러스터 1에

            - 즉, 농도(concentration)의 개념인것
                - 농도 alpha가 크면 Pi값이 0에 가깝게 되고 SBP는 많은 클러스터를 만듦
                - 반대로 농도가 낮으면 PI값이 1에 가까워지고 몇개의 클러스터만 만들어냄
            - 마지막으로 공분산 행렬을 위샤트 분포를 이용하여 샘플링
                - d와 V가 클러스터 분포를 제어함
                - 
        - 잠재변수 z에 대한 사전 지식이 Prior Probability이라는 확률분포 p(z)에 인코딩될 수 있음

## 9.2.4. 이상치 탐지와 특이치 탐지를 위한 여타 알고리즘

    - 일반적으로 샘플의 재구성 오차와 이상치의 재구성 오차를 비교하면 보통 후자가 훨씬 큼
    - 그래서 간단하고 효과적인 이상치 탐지 기법임
- Fast-MCD(minimum covariance determinant)
    - 보통 데이터셋을 정제할 때 많이 사용함
    - 일반적인 샘플이 하나의 가우시안 분포에서 생성되었다고 가정
        - 그리고 다른 가우시안 분포에서 생성된 이상치로, 이 데이터가 오염되었다고 가정
    - 그래서 알고리즘이 가우시안 분포의 파라미터를 추정할 때, 이상치로 의심되는 샘플을 무시
- Isolation Forest
    - 특히 고차원 데이터셋에서 이상치 감지를 위한 효율적인 알고리즘
    - 이 알고리즘은 무작위로 성장한 결정트리로 구성된 랜덤 포레스트를 만듦
    - 그 다음 각 노드에서 특성을 랜덤하게 선택
    - 그리고 랜덤한 임곗값을 골라 데이터셋을 둘로 나눔
    - 점차 분리되고 모든 샘플이 별도로 갈라져나감
    - 이상치는 일반적으로 다른 샘플과 멀리 떨어져있으므로, 모든 결정트리에 걸쳐서 펴ㅇ균적으로 정상 샘플과는 낮은 단계에서 분리되어 나감
- LOF(Local Outlier Factor)
    - 주어진 샘플 주위의 밀도와 이웃 주위의 밀도를 비교함
    - 이상치들은 종종 kNN보다 더 격리됨
- one-class SVM
    - 커널 SVM분류기는 모든 샘플을 고차원 공간에 매핑하고, 고차원 공간에서 선형 SVM분류기를 활용하여 두 클래스를 분리하게 됨
    - 여기서는 샘플의 클래스가 하나 -> But One-class SVM알고리즘이 원본 공간으로부터 고차원 공간에 있는 샘플을 분리함
    - 원본 공간에서는 모든 샘플을 둘러싼 작은 영역을 찾는 것에 해당함
    - 새로운 샘플이 이 영역 안에 놓이지 않는다면, 이는 이상치임
    - 조정할 파라미터가 적음
        - 1) 커널 SVM을 위한 하이퍼파라미터 하나
        - 2) 마진 하이퍼파라미터
            - 마진 : 실제 정상인 새로운 샘플을 이상치로 판단할 확률
    - 특히 고차원 데이터셋에 잘 작동하지만, 역시나 SVM이기 떄문에 대규모 데이터셋으로의 확장은 다소 어려움
