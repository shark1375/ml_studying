```python
import os
import tarfile
from six.moves import urllib
```

```python
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = os.path.join("Part2")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
#임의의 변수를 설정할 땐 이렇게 All Capital 처리하는 게 보는 사람 입장에서 좋을듯
```

```python
#github으로부터 데이터를 다운로드
def fetch_housing_data(housing_url = HOUSING_URL, housing_path = HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.mkdir(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path) #zip파일을 임의로 일단 열어제끼는게 tarfile인듯
    housing_tgz.extractall(path = housing_path) #zip파일 푸는 함수인 것 같음
    housing_tgz.close() 
```

```python
import pandas as pd

def load_housing_data(housing_path = HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

## 2.3.4. 테스트 세트 만들기
- 데이터 스누핑 편향을 막아야함
    - Data Snooping : 테스트 세트를 미리 들여다 보면, 결과에 영향을 줌
        - 따라서 시스템 오픈 시, 성능이 기대보다 떨어질 가능성이 높음

```python
#sklearn 안쓰고 직접 구현하기
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data)) #순열 알고리즘으로 랜덤추출
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size] #순열 array에서 개수로 slicing
    train_indices = shuffled_indices[test_set_size:] #반대로 slicing
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
print(len(train_set), len(test_set), len(housing))
```

- 그래서 데이터 셋이 작을 경우에는 계층적 샘플링(Stratified Sampling) 등을 진행해야함

- 프로젝트 진행 중 전문가가 중간 소득이 중간 주택 가격예측의 매우 중요 변수라고 함
- 그렇다면, 얘네가 전체 데이터 셋의 여러 소득 카테고리를 잘 대표할 필요가 있음
- 따라서 소득에 대한 카테고리 특성을 만들어줄 필요가 있음
- 각 계층을 구분해야하는데, 각 계층별로 충분한 수의 샘플이 들어있어야함
    - 너무 세분화하면 안됨!
    - 그리고 60,000 넘는 놈들도 있으므로, 얘네를 잘 처리해야함
        - 사실 계층화 시키는 가장 결정적인 이유는 얘네 때문이기도 함

```python
#진행
housing["income_cat"] = np.ceil(housing["median_income"]/1.5)
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace = True) 
```

```python
plt.hist(housing["income_cat"])
plt.show()
```

- 이제 소득 카테고리 기반의 계층적 샘플링 진행
- 사이킷런에는 Stratified Shuffle Split이 있어서 이를 이용하여 바로 진행하면 됨

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
#얘네는 test_set의 index를 만들어주는 놈들임
#일단 split이라는 객체를 만들어주고 해당 객체에 인자로 df와 set의 기준을 넣어줌
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
len(strat_train_set), len(strat_test_set), len(housing)
```

```python
housing["income_cat"].value_counts()/len(housing)*100
```

```python
strat_train_set["income_cat"].value_counts()/len(housing)*100
```

```python
strat_test_set["income_cat"].value_counts()/len(housing)*100
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis = 1, inplace = True)
```

## 2.5.1. 데이터 정제
- 우선 누락 값을 처리할 수 있는 함수를 만들자
- total_bedrooms feature에 값이 없는 것을 확인함
- 크게 세 가지 방법 적용
    - 1) 해당 구역 제거
    - 2) 전체 특성 삭제
    - 3) 대치
        - 이 때 대치를 쉽게 할 수 있도록 하는 것이 사이킷런의 imputer
        - 그냥 np.median으로 대치하지 않는 건 나중에 새로운 데이터가 들어올 때도 자동화하기 위함

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "median")
```

```python
housing_num.median().values #각 feature의 median과 일치
```

- text를 숫자로, 숫자를 One-hot Vector로 한 번에 처리할 수도 있음
- 사이킷런의 CategoricalEncoder임
- 다만 sklearn을 2.0으로 업데이트 해야함(했는데 왜 안되징...)
- 해서 아래의 코드로 대충 해결

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils import check_array
from sklearn.preprocessing import LabelEncoder
from scipy import sparse

class CategoricalEncoder(BaseEstimator, TransformerMixin):    
    def __init__(self, encoding='onehot', categories='auto', dtype=np.float64,
                 handle_unknown='error'):
        self.encoding = encoding
        self.categories = categories
        self.dtype = dtype
        self.handle_unknown = handle_unknown

    def fit(self, X, y=None):
        if self.encoding not in ['onehot', 'onehot-dense', 'ordinal']:
            template = ("encoding should be either 'onehot', 'onehot-dense' "
                        "or 'ordinal', got %s")
            raise ValueError(template % self.handle_unknown)

        if self.handle_unknown not in ['error', 'ignore']:
            template = ("handle_unknown should be either 'error' or "
                        "'ignore', got %s")
            raise ValueError(template % self.handle_unknown)

        if self.encoding == 'ordinal' and self.handle_unknown == 'ignore':
            raise ValueError("handle_unknown='ignore' is not supported for"
                             " encoding='ordinal'")
            
        X = check_array(X, dtype=np.object, accept_sparse='csc', copy=True)
        n_samples, n_features = X.shape

        self._label_encoders_ = [LabelEncoder() for _ in range(n_features)]

        for i in range(n_features):
            le = self._label_encoders_[i]
            Xi = X[:, i]
            if self.categories == 'auto':
                le.fit(Xi)
            else:
                valid_mask = np.in1d(Xi, self.categories[i])
                if not np.all(valid_mask):
                    if self.handle_unknown == 'error':
                        diff = np.unique(Xi[~valid_mask])
                        msg = ("Found unknown categories {0} in column {1}"
                               " during fit".format(diff, i))
                        raise ValueError(msg)
                le.classes_ = np.array(np.sort(self.categories[i]))

        self.categories_ = [le.classes_ for le in self._label_encoders_]

        return self
        
    def transform(self, X):        
        X = check_array(X, accept_sparse='csc', dtype=np.object, copy=True)
        n_samples, n_features = X.shape
        X_int = np.zeros_like(X, dtype=np.int)
        X_mask = np.ones_like(X, dtype=np.bool)

        for i in range(n_features):
            valid_mask = np.in1d(X[:, i], self.categories_[i])
            
            if not np.all(valid_mask):
                if self.handle_unknown == 'error':
                    diff = np.unique(X[~valid_mask, i])
                    msg = ("Found unknown categories {0} in column {1}"
                           " during transform".format(diff, i))
                    raise ValueError(msg)
            else:
                X_mask[:, i] = valid_mask
                X[:, i][~valid_mask] = self.categories_[i][0]
                
            X_int[:, i] = self._label_encoders_[i].transform(X[:, i])

        if self.encoding == 'ordinal':
            return X_int.astype(self.dtype, copy=False)

        mask = X_mask.ravel()
        n_values = [cats.shape[0] for cats in self.categories_]
        n_values = np.array([0] + n_values)
        indices = np.cumsum(n_values)

        column_indices = (X_int + indices[:-1]).ravel()[mask]
        row_indices = np.repeat(np.arange(n_samples, dtype=np.int32),
                                n_features)[mask]
        data = np.ones(n_samples * n_features)[mask]

        out = sparse.csc_matrix((data, (row_indices, column_indices)),
                                shape=(n_samples, indices[-1]),
                                dtype=self.dtype).tocsr()
        if self.encoding == 'onehot-dense':
            return out.toarray()
        else:
            return out
```

```python
#from sklearn.preprocessing import CategoricalEncoder
cat_encoder = CategoricalEncoder()
housing_cat_reshaped = housing_cat.values.reshape(-1,1)
housing_cat_1hot = cat_encoder.fit_transform(housing_cat_reshaped)
housing_cat_1hot
```

- 기본적으로 CategoricalEncoder는 Sparse Matrix를 출력하지만, Dense Matrix(일반 행렬)을 원할 경우,
- encoding 매개변수를 "onehot-dense"로 지정하면 됨

## 2.5.3. 나만의 변환기
- 내가 만든 파이프라인의 기능을 사이킷런과 매끄럽게 연동하고 싶으면, 변환기를 만드는게 좋음
- 사이킷런은 상속이 아닌 덕 타이핑을 지원하므로, fit(), transform()메서드를 구현한 클래스를 만들어 사용
    - 덕타이핑 : 상속/인터페이스 구현이 아니라 객체의 속성이나 메서드가 객체의 유형을 결정하는 것
- 만약 BestEstimator를 상속하여 클래스를 만들면, 하이퍼파라미터 튜닝에 필요한 get_params()와 set_params()를 자동으로 얻게 됨

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
            
```

## 2.5.5. 변환 Pipeline
- 아무래도 전체적으로 변환 단계가 많아질 수 밖에 없는데, 이 때 순서가 매우 중요
- 따라서 이런 연속된 변환을 순서대로 처리할 수 있어야함
- 사이킷런의 Pipeline 클래스 활용

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline([("imputer", SimpleImputer(strategy = "median")),
                        ("attribs_adder", CombinedAttributesAdder()),
                        ("std_scaler", StandardScaler())])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

- Pipeline은 연속된 단계를 나타내는 이름/추정기 쌍의 목록을 입력값으로 받음
- 다음의 주의사항을 짘야함
    - 마지막 단계에서는 변환기와 추정기를 모두 사용할 수 있음
    - 하지만, 마지막은 모두 변환기여야함(즉, fit_transform() 메서드를 가지고 있어야함!!)
    - 이중 밑줄문자 __ 는 포함해선 안됨
- 즉, pipeline객체의 fit()메서드를 호출하면, 모든 변환기의 fit_transform() 메서드를 순서대로 호출하며, 마지막 단계에서는 fit을 수행함
- 위에서는 fit_transform이었기 때문에 마지막까지 전부 fit_transform()메서드를 호출한 것

- 수치형 컬럼을 넘파이로 추출하는 대신 판다스의 DF를 파이프라인에 직접 넣을 수 있으면 좋을 것
- 아무래도 보고 정리하기가 훨씬 수월하니까
- 사이킷런이 판다스를 직접 다룰 수는 없지만, 이를 처리하는 변환기를 직접 만들 수는 있음

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y = None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
```

- 필요한 피쳐만 선택하여 DF를 NP배열로 바꾸는 식으로 데이터 변환
- 이를 통해 데이터 프레임을 받아 수치형만 다루는 파이프라인을 만들 수 있음
- 이를 파이프라인의 가장 앞 쪽에 설계하여 데이터프레임을 바로 받아 처리할 수 있음

```python
num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

num_pipeline = Pipeline([
    ("selector", DataFrameSelector(num_attribs)),
    ("imputer", SimpleImputer(strategy = "median")),
    ("attribs_adder", CombinedAttributesAdder()),
    ("std_scaler", StandardScaler()),
])
```

```python
cat_pipeline = Pipeline([
    ("selector", DataFrameSelector(cat_attribs)),
    ("cat_encoder", CategoricalEncoder(encoding = "onehot-dense")),
])
```

- 두 개로 만들어진 파이프라인을 하나로 합치려면 사이킷런의 FeatureUnion을 이용해야함
    - 변환기 목록이나 모두 변환기로 이루어진 파이프라인을 전달하고 transform() 메서드를 호출하면 각 변환기의 transform()메서드를 병렬로 실행
    - 그 다음 각 변환기의 결과를 합쳐 반환함
    - fit()을 실행하면, 각 변환기의 fit()메서드를 실행함

```python
from sklearn.pipeline import FeatureUnion
full_pipeline = FeatureUnion(transformer_list = [
    ("num_pipeline" , num_pipeline),
    ("cat_pipeline" , cat_pipeline),
])
```

```python
housing_prepared = full_pipeline.fit_transform(housing)
```

# 2.7. 모델 세부 튜닝

## 2.7.1. 그리드 탐색
- 가장 단순한 방법
- 만족할만한 Hyper Parameter(이하 HP)조합을 찾기 위해 수동으로 HP를 조정
- 지루하고 Cost가 큼
    - 이를 대체하기 위해 사이킷런의 GridSearchCV가 있음
- 탐색하고자 하는 HP와 시도해볼 값을 지정하기만 하면 됨
- 그러면 모든 HP에 대해 CV로 평가함
    - 내가 KLID때 했던 실수?를 쉽게 하는 방법이 있...

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {"n_estimators" : [3, 10, 30], "max_features" : [2, 4, 6, 8]},
    {"bootstrap" : [False], "n_estimators" : [3, 10], "max_features" : [2, 3, 4]}
]
```

- 먼저 첫 dict에 포함된 n_estimators와 max_features의 조합으로 테스트(붓스트랩은 O)
    - 그래서 결국 총 12개 조합을 테스트 하는 것
- 그 다음에는 bootstrap을 해지하고 6개 조합을 테스트
- 또한 5fold CV를 활용하므로, 18개 조합에 5회씩의 모델링+검증이 들어감
    - 따라서, 모델링을 총 90번 하는 꼴

```python
forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv = 5,
                          scoring = "neg_mean_squared_error",
                          return_train_score = True)

grid_search.fit(housing_prepared, housing_labels)
```

```python
# 최적의 HP 찾기
grid_search.best_params_
```

- n_estimator가 30인 것은, 우리가 지정해놓은 최댓값
- 따라서 더 범위를 늘려가며 테스트해봐야함
- 또한 bootstrap이 True로 디폴트가 잡혀있는 모델 -> 별도의 표시가 없으므로 True가 좋은 것으로

```python
# 최적의 모델 찾기
grid_search.best_estimator_
```

## 2.7.3. 앙상블 학습
