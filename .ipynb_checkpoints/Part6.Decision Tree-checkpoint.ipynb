# 6.1. 결정트리 학습 및 시각화

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:]
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth = 2)
tree_clf.fit(X, y)
```

- export_graphviz()함수로 그래프를 .dot파일로 출력 가능
- 이걸로 시각화 쌉 가능


```python
from sklearn.tree import export_graphviz

export_graphviz(
        tree_clf,
        out_file = "./Part6.iris_tree.dot",
        feature_names = iris.feature_names[2:],
        class_names = iris.target_names,
        rounded = True,
        filled = True
        )
```

- graphviz내의 dot 명령줄 도구로 pdf, png로 포맷 변경 가능
    - 해당 dot파일의 디렉터리로 이동해서 다음 명령어 실행
    - dot -Tpng Part6.iris_tree.dot -o Part6.iris_tree.png

# 6.2. Prediction
- 예측 방식은 아래와 같음(위 그림 참고)
    - 루트 노드(깊이가 0인 맨 꼭대기 노드)에서 시작
        - 이 노드는 꽃잎의 길이가 2.45보다 작거나 같은지를 검사함
        - True일 경우 왼쪽으로, False일 경우 오른쪽으로 분류
    - 왼쪽으로 갔으면 setosa로 분류
    - 오른쪽으로 갔으면 다시 1.75보다 작거나 같은지 검사
        - 작을 경우 versicolor로 분류
        - 클 경우 virginca로 분류      
- 노드 내 샘플 속성은 얼마나 많은 훈련 샘플이 적용되었는지 체크하는 것
    - 그게 어떻게 구성되는지 판단하는게 value속성임
- gini계수는 불순도를 측정하는 것
    - 만약 한 노드의 모든 샘플이 같은 클래스에 속할 경우 이 노드를 순수하다(gini = 0)이라고 함
    - gini impurity는 다음과 같이 계산함
        - Gi = 1 - Sigma(k =1 , n)[pi,k^2]
            - pi,k는 ith 노드의 훈련 샘플 중 클래스 k에 속한 샘플의 비율
- 가장 큰 결정 트리의 장점은 데이터 전처리가 거의 필요하지 않다는 점임
    - 스케일링도 필요없음
- 사이킷런은 이진트리만 만드는 CART알고리즘을 사용
    - 따라서 모든 노드(리프노드 제외)는 자식 노드를 두개 씩 가짐
    - But ID3같은 알고리즘은 둘 이상의 자식 노드를 가질 수도 있음
