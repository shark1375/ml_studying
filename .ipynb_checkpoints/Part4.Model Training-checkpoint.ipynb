- 킹반적으로 작동원리는 몰라도 됨
- 하지만, 디테일한 부분에서의 차이를 만들기 위해서는 반드시 필수
- 특히 ANN과 이로 인한 Deep Learning 모델들을 활용하기 위해서는 필수

# 4.1. Linear Regression
- 벡터 식으로의 변형 시, y_hat = h_theta(x) = theta'x
    - theta : theta_0 ~ theta_n까지의 가중치를 담고 있는 파라메터 벡터
        - theta_0는 bias인 beta
    - x : x_0 ~ x_n까지의 feature 벡터
        - x_0는 언제나 1
    - h_theta는 모델 파라메터 theta를 이용한 hypothesis 함수임
- 모델 훈련 = 훈련 세트에 가장 잘 맞도록 모델 파라메터를 설정하는 것
    - 이를 위해 먼저 모델이 훈련 데이터에 얼마나 잘 맞는지 측정해야함
- 회귀에 가장 널리 쓰이는 측정지표는 RMSE
    - RMSE보다는 MSE를 최소화하는 것이 더 편리
    - MSE(X, h_theta) = 1/m * Sigma(i=1, m)(theta'x_i - u_i)^2
    - 앞으로는 그냥 MSE(theta)로 표기할 것

## 4.1.1. Normal Equation
- 수학공식을 이용한 해석적인 방법으로 theta를 찾는 것
- 정규방정식은 아래와 같음
    - theta_hat = (X'X)^-1 X'y

```python
import numpy as np
import matplotlib.pyplot as plt

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1) #임의의 회귀식 만들기

plt.plot(X, y, "bo")
plt.show()
```

- 정규방정식을 이용해 theta_hat 계산

```python
X_b = np.c_[np.ones((100,1)), X] #모든 샘플에 x_theta = 1을 추가
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) #linalg는 선대모듈 -> inv는 역행렬 함수
```

```python
theta_best
```

- 4와 2를 기대했으나, 노이즈때문에 원래 함수의 파라미터를 정확하게 재현하지는 못했음

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.legend(["predict", "real"])
plt.show()
```

- 사이킷런에서 LinearRegression은 기본적으로 Normal Equation을 활용함

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
lin_reg.predict(X_new)
```

## 4.1.2. 계산 복잡도
- 정규방정식은 (n+1) by (n+1) 크기의 X'X의 역행렬을 계산함
- 역행렬 계산 복잡도는 일반적으로 O(n^2.4) ~ O(n^3) 사이
    - 즉 특성 수가 두 배로 늘어나면 계싼시간이 2^2.4 ~ 2^3배로 증가하게 됨
    - 특성 수가 많아지면 많아질 수록 계산 속도가 정말 많이 늘어난다는 것
- 다행인점은 sklearn에서 수행하는 역행렬 계산법은 O(n^2)으로 최적화 되어있으며, 차원수에서는 제곱으로 늘어나지만 훈련 세트의 샘플 수에는 선형적으로만 증가한다는 점(즉, O(m))임
- 또한, 정규방정식이나 타 알고리즘으로 학습된 선형회귀모델은 prediction이 빠르다는 장점이 있음
    - 예측 계산 복잡도는 샘플 수와 특성 수에 선형적임

# 4.2. Gradient Discent Regression
- 특성이 매우 많고 훈련 샘플이 너무 많아 메모리 문제가 생길 때 쓰기 적절한 방법임
- 여러 종류의 문제에서 최적의 해법을 찾을 수 있는 최적화 알고리즘
- 기본적인 아이디어 : 비용함수의 최소화를 위해 반복해서 파라미터를 조정해가는 것
    - 마치 짙은 안개로 길을 잃었을 때, 골짜기를 찾기 위해 가장 가파른 곳을 선택해서 가는 것과 같음
- 파라미터 벡터 theta에 대해 비용함수의 현재 그래디언트를 계산
    - 그래디언트가 감소하는 방향으로 진행(전역 최솟값 찾기)
    - 그래디언트가 0이 되는 지점이 전역 최솟값
- 구체적으로 보면 theta를 무작위 초기화하여 임의의 값으로 시작하며, 한 번에 조금씩 비용함수가 감소되는 방향으로 진행 -> 알고리즘이 최솟값에 수렴할 때까지 성능을 향상시키는 형태
- 
- 가장 중요한 하이퍼 파라메터는 학습 스텝의 크기 -> 이를 Learning Rate라고 함
    - 학습률이 너무 작으면 알고리즘이 수렴하기위한 반복을 너무 많이 필요로함
    - 학습률이 너무 크면 수렴하는 과정에서 전역 최솟값을 지나치기 쉬움
- 모든 비용함수가 2차 곡선의 형태는 아님
    - 굴곡이 심할 경우 지역 최솟값에 수렴할 가능성이 있음
- 다행히 MSE 비용 함수는 볼록 함수임
- 경사하강법을 활용할 때는 반드시 스케일을 맞춰야함
    - 그렇지 않으면 수렴하는데 굉장히 오랜 시간이 걸리게 됨
- 최적의 파라메터를 찾는 공간은 "파라미터 공간"임
    - 파라미터가 많을 수록 공간의 차원이 커져 검색이 어려워짐

## 4.2.1. 배치 경사 하강법
- 어느 방향으로 갔을 때 더 급경사인지 계산할 필요가 있음
    - 이 때 편미분(편도함수)가 필요함
    - 모든 차원에 대해 편미분을 통해서 계산함
- 즉, 파라미터 theta_j에 대한 비용하수는 dMSE(theta)/dtheta_j가 됨
- dMSE(theta)/dtheta_j = 2/m Simga(i=1, m)(theta'x(i) - y(i)x(i)_j
    - 이를 벡터 식으로 한 번에 계산하면
    - ▽_theta MSE(theta) = 2/mX'(Xtheta-y) : ▽_theta MSE(theta)는 각 방향으로 미분한 MSE의 벡터
- 이는 매 경사 하강 스텝에서 전체 훈련 세트 X에 대해 계산
    - 그래서 Batch GD라고 함
    - 그런데 이럴 경우 매 스텝에서 전체 데이터를 활용하므로, 속도가 느릴 수 밖에 없음
    - 다만 경사하강법은 feature수에 민감하지 않기 때문에, 데이터의 차원수가 너무 높을 때는 NE보다 훨씬 좋은 성능을 보임
- 만약, 근데 그래디언트 계산 과정에서 위로 향하는 기울기가 구해지면, 방향을 틀어야함
    - 즉 theta에서 ▽_theta MSE(theta)(이하 ▽)를 빼야한다는 것임
    - 이때 활용되는게 학습률 eta
    - 결국 theta_nextstep = theta - eta*▽

```python
eta = 0.1 #lr
n_iteration = 1000
m = 100

theta = np.random.randn(2,1) #theta vector 초기화

for iteration in range(n_iteration):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

```python
theta
```

- 학습률은 적절한 기울기와 theta값을 찾는 데 중요한 하이퍼 파라미터임
- 학습률이 너무 낮으면 최적점에 도달하는데 굉장히 오랜 시간이 걸릴 것
- 너무 높으면 아예 값이랑 멀어져버릴 수 있음
    - 이 때 주로 grid search를 하게 됨
        - 근데, 그리드 서치에서 수렴하는데 너무 오래 걸리는 모델이 있을 수 있음
        - 그런 모델을 방지하기 위해서는 반복 횟수를 제한할 필요가 있음
- 반복 횟수는 보통 반복 횟수를 아주 크게 지정하고, 그래디언트 벡터가 아주 작아지면
    - vector의 norm이 어떤 값 epsilon보다 작아지면 경사하강법이 최솟값에 거의 도달한 것이므로 알고리즘을 중지시킴


## 4.2.2. Stochastic GD
- 배치경사하강 -> 전체 훈련 세트를 사용하는 문제
    - 그래서 훈련세트 크기에 민감
- stochastic은 매 스텝에서 딱 한개의 샘플을 무작위로 선택하고 그 하나의 샘플에 대한 그래디언트 계산
    - 매 반복에서 매우 적은 데이터만 처리
        - 따라서 알고리즘의 속도가 훨씬 빠름
    - 또한 매 반복에서 매우 적은 메모리를 사용하므로 매우 큰 훈련세트로 훈련시킬 수 있음
- 반면, 확률적(무작위)기반 - 배치 경사하강법보다 훨씬 불안정
    - 비용함수가 최솟값에 다다를 떄까지 부드럽게 감소X -> 요동침
    - 최솟값에 안착 자체는 못할 가능성이 높으며, 좋은 파라미터를 구할 순 있지만, 최적은 아닐 것
- 무작위성은 지역 최솟값의 딜레마에서는 벗어날 수 있으나, 전역 최솟값을 궁극적으로 구할 수 없다는 단점이 있음
- 이 딜레마를 해결하기 위해서는 학습률을 점진적으로 떨어뜨릴 수 있음
    - 시작 시 학습률을 높임 -> 수렴 속도 빠르게 + 지역 최솟값 탈피
    - 점차 작게 줄여서 지역 최솟값에 수렴하도록
- 매 반복에서 이렇게 학습률을 함수를 통해 구현하여 감소시키는데 이를 "학습 스케쥴"이라고 부름
    - 학습스케쥴이 너무 빠르게 줄면, 지역 최솟값에 갇히거나, 중간에 멈출 수 있음
    - 너무 천천히 줄면, 최솟값 언저리만 계속 맴돌거나, 훈련을 너무 일찍 중지하여 지역 최솟값에 갇힐 수 있음
   

```python
n_epochs = 50
t0, t1 = 5, 50 #학습 스케쥴 파라메터

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2, 1) #무작위 초기화

for epoch in range(n_epochs):
    for i in range(m): #하나의 에포크에서 x의 개수만큼 반복시키머
        random_index = np.random.randint(m)
        xi = X_b[random_index : random_index + 1]
        yi = y[random_index : random_index + 1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta*gradients
```

```python
theta
```

- 무작위 선택 -> 어떤 샘플은 한 에포크에서 여러 번 선택될 수도 있고 어떤 샘플은 선택되지 못할 수도 있음
- 에포크 내에서 모든 샘플을 활용하게 하려면, 훈련 세트를 섞은 후 차례대로 선택 + 섞기 -> 선택 + 섞기 -> 선택 하면 되긴 하는데, 수행속도가 늦어짐
- SGDRegressor 클래스로 SGD 사용 가능

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(max_iter = 50, penalty = None, eta0 = 0.01)
sgd_reg.fit(X, y.ravel())
```

```python
sgd_reg.intercept_, sgd_reg.coef_
```

# 4.4. Learning Curve
- 고차 다항 회귀를 적용하면 1차 선형회귀에서보다 훨씬 더 훈련 데이터에 잘 맞추려고 할 것임
- 데이터셋 하나 하나에 다 맞추려고 하다보니, 엄청나게 복잡한 그래프를 그리며 접근할 것임
- 위 데이터를 그렇게 분석할 경우, 오버피팅이 발생(선형회귀는 언더피팅이 될 것)
- CV 외 다른 방식으로 오버피팅/언더피팅을 살펴보기 위해 사용하는 것이 학습곡선임
    - 훈련 세트와 검증 세트의 모델 성능을 훈련 세트 크기의 함수로 구현
    - 이를 하기 위해서는 훈련세트에서 다른 서브 세트를 만들어 내어 모델을 여러번 훈련시키면 됨

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2)
    train_errors, val_errors = [], []
    
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mse(y_train[:m], y_train_predict))
        val_errors.append(mse(y_val, y_val_predict))
        
    plt.plot(np.sqrt(train_errors), "r--", linewidth = 2, label = "training")
    plt.plot(np.sqrt(val_errors), "b--", linewidth = 3, label = "validation")
    plt.ylabel("rmse")
    plt.legend()
    plt.show()
```

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
```

- 초기에는 당연히 제대로 학습할 수 없기 떄문에 검증오차가 매우 큼
- 하지만 점점 오차가 줄어들며 그 사이가 가까워짐
- 두 곡선이 어느정도 평평하긴 하지만, 높은 오차 수준에서 평평
    - 전형적인 과소적합

```python
from sklearn.pipeline import Pipeline

polynomial_regression = Pipeline([
        ("poly_featrues", PolynomialFeatures(degree = 10, include_bias = False)), #fit_transform을 이용
        ("lin_reg", LinearRegression()) #마지막에는 fit 이용 가능
    ])
plt.ylim(0, 3)
plot_learning_curves(polynomial_regression, X, y)
```

- 전형적인 과대적합의 경우(랑은 조금 다른데...여튼)
    - 훈련 데이터의 오차가 선형회귀때보다 더 낮음
    - 두 곡선 사이의 공간이 있음(원래는 있음) -> 훈련 데이터에서의 성능이 존나 좋다 -> 즉 과대적합 되었다

- Var-Bias Trade Off
    - 편향 : 잘못된 가정으로 인한 것 -> ex) 실제 데이터는 2차인데 선형으로 가정
        - 높은 편향을 가질 경우 과소적합되기 쉬움
    - 분산 : 작은 변동에 모델이 과도하게 민감한 것
        - 자유도가 높은 모델(ex : 고차원 다항 회귀 모델)은 높은 분산을 갖기 쉬움
        - 오버피팅의 가능성이 높음
    - 줄일 수 없는 오차
        - 데이터 자체의 노이즈로 인해 발생
        - 노이즈를 제거하는 것이 유일한 방법임
    - 모델의 복잡도가 높아지면 통상적으로 Var증가, Bias 감소
    - 반대의 경우는 역시 반대로
        - 따라서 트레이드 오프라고 부름

# 4.5. Linear Model with Regulation
- 오버피팅을 방지하는 좋은 방법 중 하나는 Regulation
- 자유도를 줄여 과대적합을 방지
    - ex) 다항 회귀모델의 규제 -> 다항식의 차수를 감소시키는 것
- 선형 회귀모델에서는 보통 모델의 가중치를 제한하여 규제를 가함
    - ex) 릿지, 라쏘, 엘라스틱넷

## 4.5.1. Ridge Regression
- 규제가 추가된 선형 회귀모델
- 규제항 alpha * Sigma(i = 1, n)theta_i^2이 비용함수에 추가됨
- 학습 알고리즘을 데ㅔ이터에 맞추는 것 + 모델의 가중치가 가능한 작게 유지되도록 함
- 훈련 중에만 비용함수에 추가되며, 훈련이 끝난 후에는 모델의 성능을 규제가 없는 성능지표로 평가함
- alpha : 모델 규제량을 결정하는 하이퍼파라메터
    - alpha = 0 -> 릿지회귀 = 선형회귀
    - alpha가 커지면 모든 가중치가 거의 0에 가까워짐
- 릿지 회귀의 비용함수 : J(theta) = MSE(theta) + alpha*1/2*Sigma(i=1, n)theta_i^2
- theta_0는 규제 대상이 아님 -> 따라서 i=1부터 시작인 것
- w가 theta_1 ~ theta_n으로 구성된 벡터라고 정의하면, 규제항은 1/2(||w||_2)^2과 같음
    - 이 때 ||w||_2를 l2 norm이라고 표현함
- 이를 경사하강법에 적용할 때는 MSE 그래디언트 벡터에 alpha * w를 더하면 됨
    - 다항회귀여도 alpha를 크게 조정하면 선형의 형태를 띄게 됨
- 릿지 회귀 역시 Normal Equation과 GD로 모두 계산할 수 있음
    - 장단점은 이전의 그것과 같음
- 릿지 회귀의 정규방정식의 해 : theta_hat = (X'X + alphaA)^-1·X^r·y
    - A는 편향에 해당하는 맨 왼쪽 원소만 0이고 나머지는 대각원소가 1인 (n+1)x(n+1) 단위행렬
- 릿지회귀 적용방법은 다음과 같음

- 촐레스키 행렬분해로 정규방적식의 해를 구한 케이스

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 1, solver = "cholesky") #촐레스키 행렬분해로 릿지회귀의 해를 구하기
ridge_reg.fit(X, y)
ridge_reg.predict([[1.5]])
```

- GD를 활용한 릿지회귀 케이스

```python
sgd_reg = SGDRegressor(max_iter = 5, penalty = "l2") #l2 규제 = Ridge
sgd_reg.fit(X, y.ravel())
sgd_reg.predict([[1.5]])
```

## 4.5.2. LASSO
- Ridge에서는 l2 norm을 2로 나눈 가중치 규제를 활용
    - 여기서는 대신 가중치 벡터의 l1 norm을 활용함
- 비용함수 J(theta) = MSE(theta) + alpha * Sigma(i = 1, n)|theta_i|
- 라쏘의 가장 중요한 특징 : 가중치를 완전히 제거하려고 함
    - 즉 특정 가중치들을 0에 가까워지는게 아니라 0 자체를 만들어버리려고 함
    - 다시 말하면, 라쏘 회귀는 자동으로 feature selection을 진행하며 Sparse Model을 만듦
        - 0이 아닌 특성의 ㅡ가중치가 적음
- 라쏘 비용함수를 살펴보면 GD의 경로가 지그재그로 튐
    - 이는 theta_2 = 0에서 갑자기 기울기가 바뀌기 때문에
    - 전역 최솟값에 수렴하기 위해서는 학습률을 점전적으로 줄여가야함
- 라쏘의 비용함수는 theta_i = 0인 지점에서 미분이 불가능함
    - 하지만 thheta_i = 0일 때 서브그래디언트 벡터를 활용하면 GD를 활용할 수 있음
- 서브 그래디언트 벡터 g(theta, J) = ▽MSE(theta) + alpha(sign_vector)
    - sign_vector는 각 원소가 sign(theta_i)인 벡터
    - sign(theta_i)는 theta_i < 0 -> -1, =0 -> 0, > 0 -> 1

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.1)
lasso_reg.fit(X, y)
lasso_reg.predict([[1.5]])
```

## 4.5.3. Elastic Net
- 릿지와 라쏘를 절충한 모델
- 규제항 자체는 릿지와 회귀의 규제항을 단순히 더해서 사용
- 혼합 비율 r을 활용하여 조절
- r = 0 -> 엘라스틱넷 = 릿지회귀
- r = 1 -> 엘라스틱넷 = 라쏘회귀
- cost함수 : J(theta) = MSE(theta) + r*alpha*Sigma(i=1, n)|theta_i|+(1-r)/2*alpha*Sigma(i=1, n)theta_i^2

### 4.5.3.1. Ridge vs Lasso vs Elastic Net
- 적어도 규제가 있는 것이 대부분의 경우에 좋으므로 위 세개가 보통 Linear Reg보단 Gooooood
- 릿지가 기본
    - 하지만 실제로 쓰이는 특성이 몇 개 뿐이라고 의심이 되는 경우에는 라쏘나 엘라스틱넷 활용
- 특성 수가 훈련 샘플 수보다 많거나, 특성 몇 개가 굉장히 강한 영향력을 보일 경우에는 라쏘가 문제를 일으킬 가능성이 농후
    - 이 땐 엘라스틱 넷 활용

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha = 0.1, l1_ratio = 0.5)
elastic_net.fit(X, y)
elastic_net.predict([[1.5]])
```
